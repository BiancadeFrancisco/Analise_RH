ESSE PROJETO VISA AVALIAR A RELAÇÃO E INTERFERENCIA ENTRE O TEMPO DE EXPERIÊNCIA COM O SALÁRIO QUE ELA RECEBE.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Salary_Data.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# RENOMEAR AS COLUNAS:

df.rename(columns={
    'YearsExperience': 'Xp',
    'Salary': 'Renda'
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# VERIFICAR COLUNA RENDA COM UM GRÁFICO:

plt.figure(figsize=(10,6))
plt.title('Análise da Renda')
sns.kdeplot(df['Renda'], color = 'green');
```

```python
# VERIFICAR COLUNA EXPERIENCIA COM UM GRÁFICO:

plt.figure(figsize=(10,6))
plt.title('Análise da Experiência')
sns.kdeplot(df['Xp'], color = 'green');
```

```python
# ANALISAR A DISTRIBUIÇÃO DA RENDA:

plt.figure(figsize=(10,6))
sns.boxplot(df['Renda']);
```

```python
# ANALISAR A DISTRIBUIÇÃO DA EXPERIENCIA:

plt.figure(figsize=(10,6))
sns.boxplot(df['Xp']);
```

```python
# ANALISAR A RELAÇÃO ENTRE AS DUAS VARIÁVEIS:

plt.figure(figsize=(10,6))
plt.title('Renda X Xp')
sns.scatterplot(data = df, x='Renda', y='Xp')

# há uma regressão linear
```

```python
# ANALISAR A RELAÇÃO ENTRE AS DUAS VARIÁVEIS:

plt.figure(figsize=(10,6))
plt.title('Renda X Xp')
sns.regplot(data = df, x='Renda', y='Xp');
```

```python
#  VERIFICAR A CORRELAÇÃO ENTRE AS DUAS VARIÁVEIS:

correlacao = df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlacao, annot=True, cbar=True)
```

```python
#  VERIFICAR A CORRELAÇÃO ENTRE AS DUAS VARIÁVEIS:

correlacao = df.corr('pearson')
plt.figure(figsize=(10,6))
sns.heatmap(correlacao, annot=True, cbar=True)
```

```python
# EXISTE UMA CORRELAÇÃO FORTE ENTRE AMBAS VARIÁVEIS, 98%. QUANTO MAIOR O TEMPO DE EXPERIÊNCIA, MAIOR SERÁ O SEU SALÁRIO.
```
